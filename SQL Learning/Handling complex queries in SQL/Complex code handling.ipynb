```python
%%sql
WITH playlist_info AS 
    (
        SELECT 
            pl.playlist_id playlist_id,
            pl.name playlist_name,
            t.name track_name,
            t.milliseconds length_seconds
        FROM playlist pl
        LEFT JOIN playlist_track plt ON plt.playlist_id = pl.playlist_id
        LEFT JOIN track t ON t.track_id = plt.track_id
    )

SELECT 
    playlist_id,
    playlist_name,
    COUNT(track_name) number_of_tracks,
    (SUM((length_seconds/1000))) length_seconds
FROM playlist_info
GROUP BY playlist_id, playlist_name
ORDER BY playlist_id
```

```python
%%sql
CREATE VIEW customer_gt_90_dollars AS
    SELECT c.* FROM customer c
    INNER JOIN invoice on invoice.customer_id = c.customer_id
    GROUP BY c.customer_id
    HAVING SUM(invoice.total) > 90;
    
SELECT * FROM customer_gt_90_dollars
```

```python
%%sql
CREATE VIEW customer_usa AS 
     SELECT * FROM customer
     WHERE country = "USA";
```

```python
%%sql
SELECT * from customer_usa
limit 5;
```

```python
%%sql
SELECT * FROM customer_usa
UNION
SELECT * FROM customer_gt_90_dollars
```

```python
%%sql
WITH usa90 AS
    (
        SELECT * from customer_usa
        INTERSECT
        SELECT * from customer_gt_90_dollars
    )
    
SELECT 
    e.first_name||" "||e.last_name employee_name,
    COUNT(usa90.customer_id) customers_usa_gt_90
FROM employee e
LEFT JOIN usa90 ON e.employee_id = usa90.support_rep_id
WHERE e.title = "Sales Support Agent"
GROUP BY 1
ORDER BY 1
```

```python
%%sql
WITH
    india AS
        (
        SELECT * FROM customer
        WHERE country = 'India'
        ),
    customer_total AS
        (
        SELECT c.customer_id, SUM(iv.total) total FROM customer c
        INNER JOIN invoice iv ON c.customer_id = iv.customer_id
        GROUP BY 1
        )
        
SELECT
    i.first_name||" "||i.last_name customer_name,
    ct.total total_purchases
FROM india i
INNER JOIN customer_total ct ON ct.customer_id = i.customer_id
ORDER BY 1
```

```python
%%sql
WITH
    customer_details AS
        (
            SELECT
                customer_id,
                first_name||" "||last_name customer_name,
                country
            FROM customer
        ),
    customer_total AS
        (
            SELECT c.customer_id, SUM(iv.total) total FROM customer c
            INNER JOIN invoice iv ON c.customer_id = iv.customer_id
            GROUP BY 1
        )
SELECT 
    cd.country country,
    cd.customer_name customer_name,
    MAX(ct.total) total_purchased
FROM customer_details cd
INNER JOIN customer_total ct ON ct.customer_id = cd.customer_id
GROUP BY 1
ORDER BY 1
```
